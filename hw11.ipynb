```python
from sympy import *
import math
from sympy.plotting import plot
```

```python
x = Symbol('x')
y = Symbol('y')
```

```python
f_x = x**3 - 50*x
g_x = - x**4 + 88*x**2 - 241
```

```python
# точками пересечения будут координты, при котором будут верны уравнения f_x = g_x
```

```python
solutions = solve(f_x - g_x, (x))
```
